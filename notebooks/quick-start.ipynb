# `plit` Quick Start

1. Creating a `plot` ([link](#plot))
2. Creating a `hist` ([link](#hist))
3. Creating a `bar` ([link](#bar))
4. Combining multiple charts in one ([link](#multiple))

```python
import numpy as np
np.random.seed(1987)

# Define some data

# For the line and scatter
x1 = [np.arange(10)]
x = [np.random.random(size=(10,1)) for _ in range(4)]
y = [np.random.random(size=(10,1)) for _ in range(4)]

# For the histogram
xp = [np.random.normal(size=(100,1)), np.random.gamma(shape=1, size=(100,1)) - 2]

# For the bar chart
xg = [f"Group {i+1}"for i in range(6)]
yg = [np.random.random(size=(6)) for _ in range(2)]
```

<a id='plot'></a>

## 1.0 `plot`

For line charts and scatter plots. 

```python
from plit import plot

plot(x1, y, list("ABCD"), 'X', 'Y');
```

**By simply changing the marker type (e.g. `marker_type='o'`) you switch from line to scatter chart. See the docs for additional arguments that can be supplied as desired.**

```python
plot(x, y, list("ABCD"), 'X', 'Y', marker_type='o', grid=True);
```

**If you would like to control the colors and marker types in unison just specify `markers` using matplotlib's shorthand syntax.**

```python
plot(x1, y, list("ABCD"), 'X', 'Y', markers=['g-', 'g--', 'b-', 'b--']);
```

<a id='hist'></a>

## 2.0 `hist`

For histograms.

```python
from plit import hist

hist(xp, list("AB"), 'X', title='Histogram', bins=20);
```

<a id='bar'></a>

## 3.0 `bar`

For grouped bar charts.

```python
from plit import bar

bar(xg, yg, list("AB"),'X', 'Y', colors=list("kb"), title='Bar Chart');
```

<a id='multiple'></a>

## 4.0 Combining Multiple Charts

Combine multiple charts arbitarily by passing the axes objects as input.

```python
from plit import subplots

_, axes = subplots(nrows=2, ncols=2, figsize=(16,16))

axes[0,0] = bar(xg, yg, list("AB"), 'X', 'Y', colors=list("kb"), 
                title='Bar Chart', ax=axes[0,0])

axes[0,1] = hist(xp, list("AB"), 'X', title='Histogram', bins=20, ax=axes[0,1])

axes[1,0] = plot(x, y, list("ABCD"), 'X', 'Y', marker_type='o', 
                 grid=True, title='Scatter', ax=axes[1,0])

axes[1,1] = plot(x1, y, list("ABCD"), 'X', 'Y', title='Line', ax=axes[1,1])
```
