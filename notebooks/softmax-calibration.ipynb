## Check your model's calibration

> Warning, this example only works if you have pandas and scikit learn installed.

Use templates created with [`functools.partials`](https://docs.python.org/3.8/library/functools.html#functools.partial) or make your own. Here is an illustrative example of calculating precision and recall.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from plit import calib, subplots
save = True

def get_data():
    local_dir = '../data/processed'
    df = pd.read_csv(os.path.join(local_dir, "scaling.csv"))
    df["orig_bin"] = pd.cut(df['orig_conf'], bins=np.arange(.5, 1.05, 0.1)) 
    df["scaled_bin"] = pd.cut(df['scaled_conf'], bins=np.arange(.5, 1.05, 0.1))
    orig_acc = df.groupby("orig_bin").apply(lambda x: accuracy_score(x['obs'], x['orig_pred']))
    scaled_acc = df.groupby("scaled_bin").apply(lambda x: accuracy_score(x['obs'], x['scaled_pred']))
    return orig_acc, scaled_acc

orig_acc, scaled_acc = get_data()

fig, axes = subplots(nrows=1, ncols=2, figsize=(15,7))
expected = [0.55, 0.65, 0.75, 0.85, 0.95]

axes[0] = calib(
    y = [orig_acc.values, expected],
    labels = ['Original', 'Expected'],
    title = 'Unscaled Softmax Calibration',
    colors=['b', 'k'],
    ax=axes[0],
)

axes[1] = calib(
    y = [scaled_acc.values, expected],
    labels = ['Scaled', 'Expected'],
    title = f'Temperature Scaled Softmax Calibration',
    colors=['g', 'k'],
    ax=axes[1],
)

if save:
    fig.savefig("../figures/calibration.png", bbox_inches="tight")
```
