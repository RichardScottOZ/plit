## Creating your own templates


This shows you how to make your own template using [`functools.partials`](https://docs.python.org/2/library/functools.html#functools.partial). 

### Step 1: Create the partial

Here is a simple example that shows how to get started. Put the template in `templates.py` and create a pull request.

```python
from plit import hist
from functools import partial
import numpy as np

prob_hist = partial(
    hist,
    ylab="Observation Count (Valid)",
    xlab="Probability Bucket",
    bins=np.arange(0, 1.01, 0.05),
)
```
