```python
import numpy as np
np.random.seed(1987)

x = [np.arange(10)]
y = [np.random.random(size=(10,1)) for _ in range(4)]
```

First, lets examine the result with matplotlib. This is a VERY basic chart but it takes 9 lines of code to accomplish.

```python
import matplotlib.pyplot as plt
```

```python
colors = ["g", "b", "r", "k", "c"]
fig, ax = plt.subplots(figsize=(10, 7))
for i in range(len(y)):
    ax.plot(x[0], y[i], c=colors[i], label=list("ABCD")[i])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title", loc="left", color="k")
ax.grid(True)
ax.legend();
```

Next, lets create the same chart with a single line of plit.

```python
from plit import plot
```

```python
plot(x, y, list("ABCD"), "X", "Y", title="Title", grid=True);
```
