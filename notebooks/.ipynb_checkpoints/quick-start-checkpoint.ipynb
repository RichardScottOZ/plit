## 1.0 `plot`

For line charts and scatter plots. 

```python
from plit import plot

ax = plot(
    x_li=[np.arange(10)],
    y_li=[np.random.random(size=(10,1)) for _ in range(4)],
    labels=[f"Series {i}" for i in list("ABCD")],
    xlab='X',
    ylab='Y',
    marker_type='--'
)
```

**By simply changing the marker type (e.g. `marker_type='o'`) you switch from line to scatter chart**

```python
ax = plot(
    x_li=[np.random.random(size=(10,1)) for _ in range(4)],
    y_li=[np.random.random(size=(10,1)) for _ in range(4)],
    labels=[f"Series {i}" for i in list("ABCD")],
    xlab='X',
    ylab='Y',
    marker_type='o',
    grid=True
)
```

**If you would like to control the colors and marker types just specify `markers` using matplotlib's shorthand syntax.**

```python
ax = plot(
    x_li=[range(10)],
    y_li=[np.random.random(size=(10,1)) for _ in range(4)],
    labels=[f"Series {i}" for i in list("ABCD")],
    xlab='X',
    ylab='Y',
    markers=['g-', 'g--', 'b-', 'b--']
)
```

## 2.0 `hist`

For histograms.

```python
from plit import hist
ax = hist(
    [np.random.normal(size=(100,1)), np.random.random(size=(100,1))*4 - 2], 
    title='Histogram', 
    xlab='Probabilities',
    bins=20,
    labels=[f"Quantity {i}" for i in list("AB")]
)
```

## 3.0 `bar`

For grouped bar charts.

```python
from plit import bar

ax = bar(
    x = [f"Group {i+1}"for i in range(6)],
    y_li = [np.random.random(size=(6)) for _ in range(2)],
    labels = ['Series A', 'Series B'],
    title = 'Bar Chart',
    colors=['b', 'k'],
    xlab = 'X axis label',
    ylab = 'Y axis label',
)
```

## 4.0 Combining Multiple Charts

Combine multiple charts arbitarily by passing the axes object as input.

```python
from plit import subplots

_, axes = subplots(nrows=2, ncols=2, figsize=(16,16))

axes[0,0] = bar(
    x = [f"Group {i+1}"for i in range(6)],
    y_li = [np.random.random(size=(6)) for _ in range(2)],
    labels = ['Series A', 'Series B'],
    title = 'Bar Chart',
    colors=['b', 'k'],
    xlab = 'X axis label',
    ylab = 'Y axis label',
    ax = axes[0,0]
)

axes[0,1] = hist(
    [np.random.normal(size=(100,1)), np.random.random(size=(100,1))*4 - 2], 
    title='Histogram', 
    xlab='Probabilities',
    bins=20,
    ax = axes[0,1],
    labels=[f"Quantity {i}" for i in list("AB")]
)

axes[1,0] = plot(
    x_li = [np.random.random(size=(10,1)) for _ in range(4)],
    y_li = [np.random.random(size=(10,1)) for _ in range(4)],
    labels = [f"Series {i}" for i in list("ABCD")],
    title='Scatter Chart',
    xlab = 'X',
    ylab = 'Y',
    marker_type = 'o',
    ax = axes[1,0]
)

axes[1,1] = plot(
    x_li=[np.arange(10)],
    y_li=[np.random.random(size=(10,1)) for _ in range(4)],
    labels=[f"Series {i}" for i in list("ABCD")],
    title='Line Chart', 
    xlab='X',
    ylab='Y',
    marker_type='--',
    ax = axes[1,1]
)
```

## 5.0 ML Specific Templates

Use templates created with [`functools.partials`](https://docs.python.org/2/library/functools.html#functools.partial) or make your own. Here is an illustrative example.

**Calculate Precision and Recall on Sample Data**

```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics import recall_score, precision_score # CHANGE TO NUMPY

local_dir = '../data/processed'
d1 = pd.read_csv(os.path.join(local_dir, 'softmax-model-1.csv')) # CHANGE TO NUMPY
d2 = pd.read_csv(os.path.join(local_dir, 'softmax-model-2.csv'))

def calc_metrics_threshold(df, thresholds):
    recalls = []
    precisions = []
    for thresh in thresholds:
        df['y_pred'] = np.where(df['y_pred_proba'] > thresh, 1, 0)
        recalls.append(recall_score(df['y_true'], df['y_pred']))
        precisions.append(precision_score(df['y_true'], df['y_pred']))
    return recalls, precisions

thresholds = np.arange(0.15, 0.80, 0.01)
recalls_1, precisions_1 = calc_metrics_threshold(d1, thresholds)
recalls_2, precisions_2 = calc_metrics_threshold(d2, thresholds)
```

**Plotting Precision Recall Curve**

```python
from plit import pr_curve

ax = pr_curve(
    x_li = [thresholds],
    ylim = (0.65, 1.01),
    vref = 0.47,
    y_li=[recalls_1, precisions_1, recalls_2, precisions_2], 
    labels=['Model #1 Recall', 'Model #1 Precision', 'Model #2 Recall', 'Model #2 Precision'],
    markers = ['g-', 'b-', 'g--', 'b--'],
)
```
