## Calculate Accuracy vs. Coverage on Sample Data

> Warning, this example only works if you have pandas and scikit learn installed.

Use templates created with [`functools.partials`](https://docs.python.org/3.8/library/functools.html#functools.partial) or make your own. Here is an illustrative example of calculating precision and recall.

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score

from plit import acc_vs_cov

save = True

def get_data(filename):
    local_dir = '../data/processed'
    df = pd.read_csv(os.path.join(local_dir, filename))
    df["conf"] = np.where(df["y_pred_proba"] > 0.5, df["y_pred_proba"], 1 - df["y_pred_proba"])
    df["pred"] = np.where(df["y_pred_proba"] > 0.5, 1, 0)
    return df

def calc_metrics(filename, coverages):
    df = get_data(filename)
    metrics = []
    for coverage in coverages[0]:
        n_obs = int(coverage * df.shape[0])
        dfs = df.sort_values("conf", ascending=False).head(n_obs)
        metrics.append(accuracy_score(dfs["y_true"], dfs["pred"]))
    return metrics

coverages = [np.arange(0.10, 1.01, 0.05)]
accuracies = [
    calc_metrics("softmax-model-1.csv", coverages), 
    calc_metrics("softmax-model-2.csv", coverages),
    calc_metrics("softmax-model-3.csv", coverages)
]

acc_vs_cov(
    coverages, 
    accuracies, 
    labels = ["Model 1", "Model 2", "Model 3"], 
    ylim=[0.79, 1.01],
    subtitle = "In Domain",
    figname="acc_vs_cov.png" if save else None,
);
```
